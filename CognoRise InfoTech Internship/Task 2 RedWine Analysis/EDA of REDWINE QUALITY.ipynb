<div style="text-align: left;">
    <h3 style="font-size: 25px;">WINE ACIDIC FEATURE ANALYSIS</h3

```python
# Descriptive statistics for acidity features
display(df[['fixed acidity', 'volatile acidity', 'citric acid']].describe().T)
```

<div style="text-align: left;">
    <h4 style="font-size: 18px;">INTERPRETATION:</h4>

<div style="text-align: justified;"> 
    <p1 style="font-size: 18px;"> 
The descriptive statistics show that fixed acidity averages 8.4g/L with a standard deviation of 1.7g/L, volatile acidity averages 0.5g/L with a standard deviation of 0.2g/L, and citric acid averages 0.3g/L with a standard deviation of 0.2g/L, indicating variability in acidity levels across the wines.
    </p1>

```python
# Correlation matrix for acidity features
corr_acidity = df[['fixed acidity', 'volatile acidity', 'citric acid', 'quality']].corr().T
display(corr_acidity)

sns.heatmap(corr_acidity, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Acidity Features')
plt.show()
```

<div style="text-align: left;">
    <h4 style="font-size: 18px;">INTERPRETATION:</h4>

<div style="text-align: justified;"> 
    <p1 style="font-size: 18px;">
The correlation matrix shows how 'fixed acidity,' 'volatile acidity,' 'citric acid,' and 'quality' relate to each other. For example, 'fixed acidity' has a moderate positive correlation with 'quality'at 0.12, indicating that higher fixed acidity is somewhat associated with better quality. 'Volatile acidity' has a weak negative correlation with 'quality' at -0.39, suggesting a slight association with lower quality. 'Citric acid' has a moderate positive correlation with 'quality' at 0.23, indicating a potential link between higher citric acid and better wine quality.
    </p1>

```python
#Calculate the average values of the acidity features for each quality rating
acidity = df.groupby('quality')[['fixed acidity', 'volatile acidity', 'citric acid']].mean().reset_index()
display(acidity)

# Melt the dataframe to have a long format suitable for seaborn's barplot
acidity_melted = pd.melt(acidity, id_vars='quality', var_name='Acidity Type', value_name='Average Value')

#Visualization
plt.figure(figsize=(8, 4))
sns.barplot(x='quality', y='Average Value', hue='Acidity Type', data=acidity_melted, palette='viridis')
plt.title('Impact of  Acidity on Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Average Value')
plt.legend(title='Acidity Type')
plt.show()
```

<div style="text-align: left;">
    <h4 style="font-size: 18px;">INTERPRETATION:</h4>

<div style="text-align: justified;"> 
    <p1 style="font-size: 18px;">
The analysis calculates the average levels of 'acidty features' for each wine quality rating. The visual further shows that higher quality wines generally have higher average fixed acidity and citric acid at 8.0g/L and 0.4g/L respectively, while lower quality wines have higher volatile acidity at 0.6g/L for wine quality rating '3'. This suggests that wines with higher fixed acidity and citric acid, but lower volatile acidity, tend to have better quality.
    </p1>

```python
# Calculate the average quality rating for each pH value and select the top 5 pH values
HighestQuality = df.groupby('quality')['pH'].mean().sort_values(ascending=False).reset_index().head(5)
display(HighestQuality)

# Custom function to format the labels
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0, 2))
        return f'{pct:.1f}%\n{val:.2f}'
    return my_autopct

# Visualization using Pandas
# Create a pie chart
plt.figure(figsize=(8, 4))
plt.pie(HighestQuality['quality'], labels=None, autopct=make_autopct(HighestQuality['quality']), 
        colors=sns.color_palette('viridis', 5), startangle=90)
plt.title('Top 5 Quality Wines by pH')
# Create the legend
pH_labels = [f'{pH:.3f}' for pH in HighestQuality['pH']]
plt.legend(pH_labels, title='pH Values', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

<div style="text-align: left;">
    <h4 style="font-size: 18px;">INTERPRETATION:</h4>

<div style="text-align: justified;"> 
    <p1 style="font-size: 18px;">

The analysis identifies the top 5 wine quality ratings based on average pH values. The pH values, which measure the overall acidity of the wine, indicate that higher quality wines tend to have specific pH levels. For example, wines with a quality rating of 8 might have an average pH of 3.50, reflecting optimal acidity. The pie chart highlights how certain pH ranges correlate with top-quality wines, suggesting that maintaining specific pH levels is crucial for producing high-quality wines.
</p1>

<div style="text-align: left;">
    <h3 style="font-size: 20px;">OBSERVATION</h3

<div style="text-align: justified;"> 
    <p1 style="font-size: 18px;">
The acidity features analysis reveals that higher quality wines typically have higher fixed acidity and citric acid, along with lower volatile acidity. Additionally, the analysis further indicates that top-quality wines have pH values around 3.50, showing an optimal acidity level. Hence this suggests that a balanced acidity profile is key to achieving higher wine quality.
    </p1>

```python
# Correlation Matrix for taste features
corr_taste= df[['residual sugar', 'chlorides', 'quality']].corr().T
display(corr_taste)

sns.heatmap(corr_taste, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Taste Features')
plt.show()
```

```python
#Calculate the average values of the taste features for each quality rating
taste = df.groupby('quality')[['residual sugar', 'chlorides']].mean().reset_index()
display(taste)

# Melt the dataframe to have a long format suitable for seaborn's barplot
taste_melted = pd.melt(taste, id_vars='quality', var_name='Taste', value_name='Average Value')

#Visualization
plt.figure(figsize=(8, 4))
sns.barplot(x='quality', y='Average Value', hue='Taste', data=taste_melted, palette='viridis')
plt.title('Impact of Taste on Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Average Value')
plt.legend(title='Taste')
plt.show()
```

The  visual further illustrate that higher quality wines generally have lower average residual sugar for instance, 2.7g/L for wine quality '3' and 2.5g/L for  wine quality '8' and for chlorides, 0.1g/L for  wine quality '3' and 0.06g/L for wine quality '8'. This prectically suggests that better quality wines tend to have lower levels of these taste features, indicating a potential link between reduced sugar and chloride levels and higher wine quality.

```python
HighestQuality = df.groupby('quality')['density'].mean().sort_values(ascending = False).reset_index()[0:5]
display(HighestQuality)

# Custom function to format the labels
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0, 2))
        return f'{pct:.1f}%\n{val:.2f}'
    return my_autopct

# Visualization using Pandas
# Create a pie chart
plt.figure(figsize=(8, 4))
plt.pie(HighestQuality['quality'], labels=None, autopct=make_autopct(HighestQuality['quality']), 
        colors=sns.color_palette('viridis', 5), startangle=90)
plt.title('Top 5 Quality Wines by density')
# Create the legend
density_labels = [f'{density:.4f}' for density in HighestQuality['density']]
plt.legend(density_labels, title='density values', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

The associated visual illustrates that these optimal density levels are closely associated with the highest quality wines, emphasizing the importance of achieving the right density for superior taste.

<div style="text-align: left;">
    <h3 style="font-size: 25px;">WINE PRESERVATIVE FEATURE ANALYSIS</h3

```python
# Descriptive Statistics presevative analysis
preservative_descriptive = df[['free sulfur dioxide', 'total sulfur dioxide']].describe().T
display(preservative_descriptive)
```

<div style="text-align: left;">
    <h4 style="font-size: 18px;">INTERPRETATION:</h4>

<div style="text-align: justified;"> 
    <p1 style="font-size: 18px;">The descriptive statistics show that 'free sulfur dioxide' has a mean of around 15mg/L and 'total sulfur dioxide' a mean of 46mg/L, with respective standard deviations of 10mg/L and 32mg/L. within the minimal and maximum range between 1.0mg/L - 72.0mg/L for 'free sulfur dioxide', and 6.0mg/L - 289 mg/L for the 'total sulfur dioxide' .This indicates that 'total sulfur dioxide' values are more spread out, reflecting greater variability in preservative use.
    </p1>

```python
# Correlation Matrix for preservative features
corr_preservative= df[['free sulfur dioxide', 'total sulfur dioxide', 'quality']].corr().T
display(corr_preservative)

sns.heatmap(corr_preservative, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Presevative Features')
plt.show()
```

<div style="text-align: left;">
    <h4 style="font-size: 18px;">INTERPRETATION:</h4>

<div style="text-align: justified;"> 
    <p1 style="font-size: 18px;">
The correlation matrix shows how 'free sulfur dioxide,' 'total sulfur dioxide,' and 'quality' relate to each other. The matrix reveals that 'free sulfur dioxide' and 'total sulfur dioxide' have a strong positive correlation, at 0.67, indicating that as one increases, the other does as well. 'Quality' has a weaker negative correlation with both sulfur dioxide measures, at -0.19 with 'total sulfur dioxide', suggesting that higher sulfur dioxide levels are somewhat associated with lower quality ratings. The heatmap visually confirms these relationships, with warmer colors indicating stronger correlations.
    </p1>

```python
#Calculate the average values of the preservative features for each quality rating
preservative = df.groupby('quality')[['free sulfur dioxide', 'total sulfur dioxide']].mean().reset_index()
display(preservative)

# Melt the dataframe to have a long format suitable for seaborn's barplot
preservative_melted = pd.melt(preservative, id_vars='quality', var_name='Preservative', value_name='Average Value')

#Visualization
plt.figure(figsize=(8, 6))
sns.barplot(x='quality', y='Average Value', hue='Preservative', data=preservative_melted, palette='viridis')
plt.title('Impact of Preservatives on Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Average Value')
plt.legend(title='Preservative')
plt.show()
```

<div style="text-align: left;">
    <h4 style="font-size: 18px;">INTERPRETATION:</h4>

<div style="text-align: justified;"> 
    <p1 style="font-size: 18px;">
The analysis calculates the average levels of 'free sulfur dioxide' and 'total sulfur dioxide' for each wine quality rating. For instance, wines rated as '5' have an average of 30 mg/L for free sulfur dioxide and 120 mg/L for total sulfur dioxide, while wines rated as '8' have lower averages of 10 mg/L and 80 mg/L, respectively. The visualization further indicates that higher quality wines generally have lower average preservative levels, which suggest that better quality wines tend to use fewer preservatives.
    </p1>
