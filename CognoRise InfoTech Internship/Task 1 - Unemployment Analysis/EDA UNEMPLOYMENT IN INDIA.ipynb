<div style="text-align: left;">
    <h1 style="font-size: 20px;">CORRELATION ANALYSIS</h1>

```python
# Compute the correlation matrix for the key variables
Correlation_Matrix = df_1[['Unemployment_Rate(%)', 'Labour_Participation_Rate(%)']].corr(numeric_only =True)

# Display the correlation matrix
display(Correlation_Matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap to visualize the correlation matrix for key metrics
plt.figure(figsize=(4, 3))
sns.heatmap(Correlation_Matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
OBSERVATION: From the Correlation matrix above its clearly evident that the unemployment rate(%) does not co-depend on the labour participation rate(%)
```

<div style="text-align: left;">
    <h1 style="font-size: 20px;">REGIONAL ANALYSIS</h1>

```python
# Comparative Analysis by region 

# Group by 'Region' and calculate mean values for 'Unemployment_Rate(%)' and 'Labour_Participation_Rate(%)'
comparative_stats = df_1.groupby('Region')[['Unemployment_Rate(%)', 'Labour_Participation_Rate(%)']].mean().reset_index()
display(comparative_stats)

# Melt the DataFrame to long format for easier plotting
df_melted = comparative_stats.melt(id_vars=['Region'], value_vars=['Unemployment_Rate(%)', 'Labour_Participation_Rate(%)'], 
                      var_name='Metric', value_name='Value')

# Visualize the comparative analysis
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Value', hue='Metric', data=df_melted, palette='viridis')
plt.title('Comparative Analysis of Unemployment Rate and Labour Participation Rate by Region')
plt.xlabel('Region')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.legend(title='Metric')
plt.tight_layout()
plt.show()
```

```python
#3 Top 5 (five ) Region with Highest Unemployment Rate
Highest_Unemployment_rate = df_1['Unemployment_Rate(%)'].groupby(by=df_1['Region']).mean().sort_values(ascending=False).reset_index()[0:5]

# Display the results
print("Top 5 Regions With Highest Unemployement_Rate(%)")
display(Highest_Unemployment_rate)
```

```python
# Top 5 (five ) Region with Highest Unemployment Rate
Highest_Unemployment_rate = df_1['Unemployment_Rate(%)'].groupby(by=df_1['Region']).mean().sort_values(ascending=False).reset_index()[0:5]

# Display the results
print("Top 5 Regions With Highest Unemployement_Rate(%)")
display(Highest_Unemployment_rate)

# Visualization using Seaborn

plt.figure(figsize=(6, 4))
sns.barplot(x='Region', y='Unemployment_Rate(%)', data= Highest_Unemployment_rate, palette='viridis', hue='Region', dodge=False)
plt.xlabel('Region')
plt.ylabel('Unemployment Rate(%)')
plt.title('Top 5 Regions with the Highest Unemployment Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
#7 Average Unemployment Rate by Month

#convert date to datetime
df_1['Date'] = pd.to_datetime(df_1['Date'], dayfirst=False)

#Extract Month from date
df_1['Month'] = df_1['Date'].dt.month_name()

#Dispaly new Colunm Name on Dataset
df_1.head()
```

```python
# Calculate the Highest Average of Unemployment rate (%) by Month
Highest_UmR_by_Month = df_1.groupby('Month')['Unemployment_Rate(%)'].mean().reset_index()
display(Highest_UmR_by_Month)
```

```python
# Visualization Using Seaborn

plt.figure(figsize=(8, 4))
sns.lineplot(x='Month', y='Unemployment_Rate(%)', data=Highest_UmR_by_Month, marker='o')
plt.xlabel('Month')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Monthly Trend of Unemployment Rate(%)')
plt.xticks(rotation=45)
plt.show()
```
